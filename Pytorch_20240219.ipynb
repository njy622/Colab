### rand
- 균등분포
- 0~1 랜덤하게 정해짐
- 균등하게 쪼개서 랜덤하게 정해줌
- y값이 없이 x값만 나옴

### randn
- -1~1까지
- 보이는건 x값만 보이지만 y값도 존재함

## rand
: 0에서 1사이 균일한 확률 분포로 실수와 난수를 만들어주는 서브 패키지 입니다.

괄호 안에 들어가는 숫자(인수)는 난수의 크기(n개)로 보시면 됩니다.


## randn
: 표준편자가 1인 가우시안의 표준 정교 분포를 따른 난수 생성 패키지입니다.

## randint
: a에서 시작하여 b-1까지 난수를 발생시키며, 괄호 안은 행렬의 size를 말합니다.

사용법은

np.random.randint(a, b, (행, 열))

```python
np.random.random(3)
```

```python
np.random.randn(3)
```

```python
np.random.random((3,3))
```

```python
# 인덱싱
p = torch.tensor([[1, 2],
                  [3, 4],
                  [5, 6]])
q = p[1, 0] #2번째 행에서 첫번째열 위치
print(q)
```

```python
p.shape
```

```python
# 슬라이싱
p = torch.tensor([[1, 2],
                  [3, 4],
                  [5, 6]])
q = p[1:2, 0:1]
print(q)
```

```python
q.shape
```

```python
# 인플레이스 연산 예제
# add
a = torch.tensor([1, 2, 3])
b = a
a.add(torch.tensor([1, 1, 1]))
print(a)
print(b)
```

```python
# add_
a = torch.tensor([1, 2, 3])
b = a
a.add_(torch.tensor([1, 1, 1]))
print(a)
print(b)
```
